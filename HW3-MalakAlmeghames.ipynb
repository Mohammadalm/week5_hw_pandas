```python
df1 = pd.melt(df_deaths_clean, id_vars=['month',"Region"],value_vars=["Males","Females"],
             var_name="Sex", value_name="value")
df1 #دمجت عمود النساء والرجال في عمود واحد سميته الجنس
```

```python
df2 = pd.melt(df_deaths_clean, id_vars=['month',"Region"],value_vars=["inside_city","outside_city"],
             var_name="Location", value_name="value")
df2 #دمجت عمود داخل المدينة وخارجها في عمود واحد سميته الموقع
```

```python
df3 = pd.melt(df_deaths_clean, id_vars=['month',"Region"],value_vars=["saudi","not_saudi"],
             var_name="Nationality", value_name="value")
df3 #دمجت عمود السعودي وغير السعودي في عمود واحد سميته الجنسية
```

```python
df_merge1 = pd.merge(df1,df2, left_index=True ,right_index=True, how="left")
df_merge1 #دمجت جدولين
```

```python
df_merge2 = pd.merge(df3,df_merge1, left_index=True ,right_index=True, how="left")
df_merge2 #دمجت جدولين
```

```python
result_df = df_merge2.drop(["month_x","month_y","Region_x","Region_y"], axis=1)
result_df #اسقطت العواميد المتشابهة في الجدول
```

```python
import seaborn as sns
sns.scatterplot(x="value_x", y="value_y", data=result_df)
#قارنت بين نسبة الوفايات بالنسبة للجنس 
#قارنت بين نسبة الوفايات بالنسبة للموقع
```
